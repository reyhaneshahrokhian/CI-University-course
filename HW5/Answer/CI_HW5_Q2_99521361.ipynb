```python
import skfuzzy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

# data1

```python
data1 = pd.read_csv('/content/data1.csv')
```

```python
# Normalizing
data1[['X', 'Y']] = StandardScaler().fit_transform(data1[['X', 'Y']])
```

```python
finalC=2
finalFPC=0

for c in range(2, 11):
    # FCM algorithm
    cntr, u, u0, d, jm, p, fpc = skfuzzy.cluster.cmeans(data1.T, c, 2, error=0.005, maxiter=1000, init=None)

    cluster_membership = np.argmax(u, axis=0)

    fig, ax = plt.subplots()

    for i in range(c):
        ax.plot(data1['X'][cluster_membership == i], data1['Y'][cluster_membership == i], 'o', label=f'Cluster {i + 1}')

    for cluster_center in cntr:
        ax.plot(cluster_center[0], cluster_center[1], 'X', markersize=10, markeredgewidth=2, label='Cluster Center')

    if fpc > finalFPC:
        finalFPC = fpc
        finalC = c


    # Plot for each cluster
    ax.set_title(f'FCM with {c} clusters')
    ax.legend()
    plt.show()
```

```python
print(f"the best fpc is {finalFPC} which is with {finalC} clustring")
```
